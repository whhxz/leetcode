```python
from typing import List


class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        numsd = dict()
        left = set()
        right = set()
        for num in nums:
            numsd.setdefault(num, 0)
            numsd[num] += 1
            if num < 0:
                left.add(num)
            else:
                right.add(num)
        res = set()
        for ln in left:
            numsd[ln] -= 1
            for rn in right:
                numsd[rn] -= 1
                t1 = target - ln - rn
                for lln in left:
                    if numsd[lln] > 0:
                        numsd[lln] -= 1
                        t2 = t1 - lln
                        if (t2 in left or t2 in right) and numsd[t2] > 0:
                            res.add(tuple(sorted([ln, rn, lln, t2])))
                        numsd[lln] += 1
                for rrn in right:
                    if numsd[rrn] > 0:
                        numsd[rrn] -= 1
                        t2 = t1 - rrn
                        if t2 in right and numsd[t2] > 0:
                            res.add(tuple(sorted([ln, rn, rrn, t2])))
                        numsd[rrn] += 1
                numsd[rn] += 1
            numsd[ln] += 1
        for r in right:
            numsd[r] -= 1
            for rr in right:
                if numsd[rr] == 0:
                    continue
                numsd[rr] -= 1
                t1 = target - rr - r
                for rrr in right:
                    if numsd[rrr] == 0:
                        continue
                    numsd[rrr] -= 1
                    t2 = t1 - rrr
                    if t2 in numsd and numsd[t2] > 0:
                        res.add(tuple(sorted([r, rr, rrr, t2])))
                    numsd[rrr] += 1
                numsd[rr] += 1
            numsd[r] += 1
        for r in left:
            numsd[r] -= 1
            for rr in left:
                if numsd[rr] == 0:
                    continue
                numsd[rr] -= 1
                t1 = target - rr - r
                for rrr in left:
                    if numsd[rrr] == 0:
                        continue
                    numsd[rrr] -= 1
                    t2 = t1 - rrr
                    if t2 in numsd and numsd[t2] > 0:
                        res.add(tuple(sorted([r, rr, rrr, t2])))
                    numsd[rrr] += 1
                numsd[rr] += 1
            numsd[r] += 1

        return [[n[0], n[1], n[2], n[3]] for n in res]
```
