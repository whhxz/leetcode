```python
class Solution:
    #这个代码写在Python里面超时，但是用Java写，提交却并没有超时，而且超过70%，不知道计算原理是什么
    def countRangeSum(self, nums, lower, upper):
        """
        :type nums: List[int]
        :type lower: int
        :type upper: int
        :rtype: int
        """
        total = 0
        sums = []
        #计算每个数到第0个的和
        #这样如果要计算指定范围的值如i~j的和就可以通过sums[j]-sums[i-1]计算得到
        count = 0
        for num in nums:
            total += num
            if total >= lower and total <= upper:
                count += 1
            sums.append(total)
        #通过并归排序计算总值
        count += self.merge(sums, len(sums) * [0], 0, len(sums) - 1, lower, upper)
        return count
    def merge(self, nums, tmps, start, end, lower, upper):
        """
        使用并归排序，是因为并归在合并前，右边数组是没有变动的，右边依次减去左边的数据，就可以计算出差值是否在访问内
        在计算完毕后在进行合并
        """
        if end <= start:
            return 0
        min_i = (end + start) >> 1
        start1, start2, end1, end2 = start, min_i + 1, min_i, end
        count = self.merge(nums, tmps, start1, end1, lower, upper) + self.merge(nums, tmps, start2, end2, lower, upper)
        ti = start
        for i in range(start, end + 1):
            if start1 > end1:
                # 这里表示需要提前右边数据，右边的数据未移动过，遍历右边减去左边所有数据，判断是否在区间内
                for j in range(start, min_i+ 1):
                    nn = nums[start2] - nums[j]
                    if nn >= lower and nn <= upper:
                        count += 1
                    elif nn < lower:
                        break
                tmps[i] = nums[start2]
                start2 += 1
            elif start2 > end2:
                tmps[i] = nums[start1]
                start1 += 1
            elif nums[start1] <= nums[start2]:
                tmps[i] = nums[start1]
                start1 += 1
            else:
                for j in range(start, min_i+ 1):
                    nn = nums[start2] - nums[j]
                    if nn >= lower and nn <= upper:
                        count += 1
                    elif nn < lower:
                        break

                tmps[i] = nums[start2]
                start2 += 1
        for i in range(start, end + 1):
            nums[i] = tmps[i]
        return count
        
```
