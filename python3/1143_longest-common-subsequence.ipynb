```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        tl1 = list(text1)
        tl2 = list(text2)
        dp = [[0] * (len(tl2) + 1) for t in range(len(tl1) + 1)]
        
        for i1, c1 in enumerate(tl1):
            for i2,c2 in enumerate(tl2):
                if c1 == c2:
                    dp[i1+1][i2+1] = max(dp[i1][i2] + 1, dp[i1+1][i2])
                else:
                    dp[i1+1][i2+1] = max(dp[i1][i2+1], dp[i1+1][i2])
        return dp[-1][-1]
                
```
